# Read SDSS data

```python
import os, sys
#import glob
#from astropy.io import fits
import matplotlib.pyplot as plt
#from subprocess import Popen, run
import h5py
import numpy as np
#import matplotlib.pylab as plt
```

```python
f_train = h5py.File("data/sdss_galaxy_spec.hdf5", "r")

# Fiber identification
plateid = f_train['raw']['plateid'][()]
fiberid = f_train['raw']['fiberid'][()]
mjd = f_train['raw']['mjd'][()]

# Position
ra = f_train['raw']['ra'][()]
dec = f_train['raw']['dec'][()]
z = f_train['raw']['z'][()]
zerr = f_train['raw']['zerr'][()]

# Physical properties
age = f_train['raw']['age'][()]
metallicity = f_train['raw']['metallicity'][()]
smass = f_train['raw']['smass'][()]

# Spectra
wavelength = f_train['raw']['wavelength'][()]

specs = f_train['raw']['spec'][()]
specerrs = f_train['raw']['specerr'][()]
```

```python
ith = 1000
spec = specs[ith]
```

```python
plt.plot(wavelength, spec)
plt.xlabel("Wavelength (Angstroms)")
plt.ylabel(r'$f_\lambda~(10^{-17}~erg~cm^{-2}~s^{-1}~\AA^{-1})$')
```
